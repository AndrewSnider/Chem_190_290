A bunch of helpful plotting examples in Python!

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib inline
```

```python
# line plot
x = np.array([1, 2, 3, 4])
y = np.square(x)
plt.plot(x, y)
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

```python
# scatter plot
x = np.array([1, 2, 3, 4])
y = np.square(x)
plt.scatter(x, y)
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

```python
# plot with error bars
x = np.array([1, 2, 3, 4])
y = np.square(x)
np.random.seed(19680801) # Fixing random state for reproducibility
yerr = np.random.rand(len(x))*2 # define errors
plt.errorbar(x, y, yerr=yerr, fmt='-o') 
plt.show()
```

```python
# log plot
x = np.linspace(1, 10, 100)
y = np.log(x)
plt.semilogy(x, y)
plt.show()
```

```python
# area under the curve
x = np.array([1, 2, 3, 4])
y = np.square(x)
plt.fill_between(x, y, color='#FF7F7F')
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

```python
# bar plot
x = np.array([1, 2, 3, 4])
y = np.square(x)
plt.bar(x, y)
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

```python
# stacked bar plot
A = [5, 15, 22]
B = [3, 8, 14]
barwidth = 0.5 
plt.bar(range(len(A)), A, color='b', edgecolor='grey', width=barwidth) # plot bar plot for dataset A
plt.bar(range(len(B)), B, bottom=A, color='r', edgecolor='grey', width=barwidth) # plot bar plot for dataset B with A as the bottom
plt.show()
```

```python
# histogram
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.show()
```

```python
# pie plot
sizes = [15, 30, 45, 10] # define the size of each piece
plt.pie(sizes)
plt.show()
```

```python
# box plot
data = np.random.rand(10, 4) # generate 10 rows and 4 columns of r.n.
plt.boxplot(data)
plt.show()
```

```python
# violin plot
data = np.random.rand(10, 4)
sns.violinplot(data=data)
plt.show()
```

```python
# heat map
data = np.random.rand(10, 12) # generate 10 rows and 12 columns of r.n.
sns.heatmap(data)
plt.show()
```

```python
# spider plot
categories = ['A', 'B', 'C', 'D']
values = [4, 3, 2, 5]
angles = [n / float(len(categories)) * 2 * np.pi for n in range(len(categories))] # compute the angle of each category
values += values[:1] # close the shape, last point is the first point
angles += angles[:1] # close the shape
ax = plt.subplot(111, polar=True) # define polar axis (note that plt.subplot(111) is the same as plt.subplot(1,1,1))
plt.xticks(angles[:-1], categories) # define x axis ticks
ax.plot(angles, values) # make the radar plot
ax.fill(angles, values, 'teal', alpha=0.1) # fill the color
plt.show()
```

```python
# radar plot
num_vars = 6 # define the number of variables
angles = np.linspace(0, 2*np.pi, num_vars, endpoint=False).tolist() # compute angles
stats = np.random.rand(num_vars) 
angles += angles[:1]
stats = np.concatenate((stats, [stats[0]])) # close the shape
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True)) # create the polar subplot
ax.fill(angles, stats, color='red', alpha=0.25) # fill the color
ax.plot(angles, stats, color='red', linewidth=2) # draw the radar lines
plt.show()
```

```python
# 3D scatter plot
# mplot3d toolkit: https://matplotlib.org/stable/users/explain/toolkits/mplot3d.html#mplot3d
# from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure() # create a new figure
ax = fig.add_subplot(111, projection='3d') # add a 3D subplot
x = np.array([1, 2, 3, 4])
y = np.square(x)
z = np.power(x, 3)
ax.set_xlabel('x value')
ax.set_ylabel('y value')
ax.set_zlabel('z value')
ax.scatter(x, y, z)
plt.show()
```

```python
# surface plot in 3D
# mplot3d toolkit: https://matplotlib.org/stable/users/explain/toolkits/mplot3d.html#mplot3d
# from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm # load the cm module from matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y) # create the mesh grid
r = np.sqrt(x**2 + y**2) # compute the distance between each grid and the origin
z = np.sin(r)
ax.set_xlabel('x value')
ax.set_ylabel('y value')
ax.set_zlabel('z value')
ax.plot_surface(x, y, z, cmap=cm.coolwarm) # create the surface in 3D, and add color
plt.show()
```

```python
# contour plot
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y) # create mesh
z = np.sqrt(x**2 + y**2) 
contour = plt.contourf(x, y, z) # filled contour
plt.colorbar(contour)  # Add a colorbar to a plot
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('square')
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.square(x)

# create 1 figure and 1 subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# create 1 figure and 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

plt.show()
```
